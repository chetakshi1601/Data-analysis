# ðŸ›’ E-Commerce Sales Analysis Project (Python + Power BI)
This project analyzes sales, profit, and discount data from an e-commerce business using Python.
We'll perform data cleaning, exploratory analysis, and visualize trends to extract key insights.

```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use('ggplot')
```

```python
# Step 2: Load Dataset
df = pd.read_csv('ecommerce_sales.csv')
df.head()
```

```python
# Step 3: Basic Info and Cleaning
df.info()
df.drop_duplicates(inplace=True)
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df.isnull().sum()
```

```python
# Step 4: Key Metrics
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
top_region = df.groupby('Region')['Sales'].sum().idxmax()
print(f'Total Sales: â‚¹{total_sales:,.2f}')
print(f'Total Profit: â‚¹{total_profit:,.2f}')
print(f'Top Region by Sales: {top_region}')
```

```python
# Step 5: Category-wise Sales
cat_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
cat_sales.plot(kind='bar', color='skyblue', title='Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Step 6: Monthly Sales Trend
df['Month'] = df['Order_Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Trend', figsize=(10,5))
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
# Step 7: Discount vs Profit Relationship
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Discount vs Profit')
plt.show()
```

```python
# Step 8: Top 10 Products by Sales
top_products = df.groupby('Product_Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='barh', color='lightgreen', title='Top 10 Products by Sales', figsize=(8,5))
plt.xlabel('Sales')
plt.ylabel('Product Name')
plt.show()
```
